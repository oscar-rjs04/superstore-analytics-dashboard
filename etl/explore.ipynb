```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../database/superstore.db")

tables = ["customers", "products", "locations", "dates", "sales"]
for t in tables:
    df = pd.read_sql(f"SELECT * FROM {t} LIMIT 3", conn)
    print(f"\n--- {t} ({pd.read_sql(f'SELECT COUNT(*) as n FROM {t}', conn).iloc[0,0]} filas) ---")
    print(df.to_string())

conn.close()
```

```python
min_order =df["Order Date"].min()
min_ship = df["Ship Date"].min()

max_order = df["Order Date"].max()
max_ship = df["Ship Date"].max()

print(f"Fecha mínima de pedido: {min_order}")
print(f"Fecha mínima de envío: {min_ship}")
print(f"Fecha máxima de pedido: {max_order}")
print(f"Fecha máxima de envío: {max_ship}")
```

```python
import sqlite3
conn = sqlite3.connect("../database/superstore.db")

# Suppliers
print(pd.read_sql("SELECT * FROM suppliers", conn))

# Sample purchases
print(pd.read_sql("""
    SELECT p.*, s.supplier_name, pr.category 
    FROM purchases p
    JOIN suppliers s ON p.supplier_id = s.supplier_id
    JOIN products pr ON p.product_id = pr.product_id
    LIMIT 5
""", conn).to_string())

# Sample inventory con alerta de reorder
print(pd.read_sql("""
    SELECT i.*, pr.category
    FROM inventory i
    JOIN products pr ON i.product_id = pr.product_id
    WHERE i.stock_qty <= i.reorder_point
    LIMIT 5
""", conn).to_string())

conn.close()
```

```python
conn = sqlite3.connect("../database/superstore.db")

print(pd.read_sql("""
    SELECT stock_qty, COUNT(*) as productos
    FROM inventory
    WHERE snapshot_date = '2014-01-01'
    GROUP BY stock_qty
    ORDER BY stock_qty
    LIMIT 10
""", conn))

# Cuantos productos tienen ventas antes de 2014?
print(pd.read_sql("""
    SELECT COUNT(DISTINCT product_id) as productos_con_ventas_2014
    FROM sales
    WHERE order_date < '2015-01-01'
""", conn))

conn.close()
```
